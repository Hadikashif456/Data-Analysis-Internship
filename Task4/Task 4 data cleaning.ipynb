# ***Data Cleaning***

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
# Drop the first row (contains question text, not actual data)
df = df.drop(0).reset_index(drop=True)
```

```python
# Basic cleaning strip spaces from column names
df.columns = df.columns.str.strip()
```

```python
# Drop columns with more than 80% missing data to simplify analysis
threshold = 0.8 * len(df)
df = df.dropna(axis=1, thresh=threshold)
```

```python
# Clean characters
df["Q4"] = df["Q4"].str.replace("â€™", "'", regex=False)
```

```python
# Rename Columns
df.rename(columns={
    "-": "Year",
    "Time from Start to Finish (seconds)": "Response_Time",
    "Q1": "Age_Range",
    "Q2": "Gender",
    "Q3": "Country",
    "Q4": "Education",
    "Q5": "Job_Title",
    "Q6": "Coding_Experience",
    "Q8": "Recommended_Language"
}, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.isnull().sum() / len(df) * 100
```

```python
# Fill Missing Values
# Fill categorical columns with "Unknown"
for col in df.columns:
    if df[col].dtype == 'object':  # categorical / text column
        df[col] = df[col].fillna("Unknown")
    else:
# Fill numeric columns with median (less sensitive to outliers)
        df[col] = df[col].fillna(df[col].median())
```

```python
df.isnull().sum()
```

```python
# Map Education Levels to Numeric Codes
education_map = {
    "I did not complete any formal education past high school": 0,
    "No formal education past high school": 0,
    "Some college/university study without earning a bachelor's degree": 1,
    "Bachelor's degree": 2,
    "Master's degree": 3,
    "Professional degree": 4,
    "Professional doctorate": 5,
    "Doctoral degree": 6,
    "I prefer not to answer": -1,
    "Unknown": -1
}

df["Education_Encoded"] = df["Education"].map(education_map).fillna(-1)

# Show a Clean Summary 
print("✅ Education Column Encoded Successfully!\n")

# Unique mappings for quick check
print("🔎 Unique Education → Encoded Mapping:")
print(df[["Education", "Education_Encoded"]].drop_duplicates().sort_values("Education_Encoded"))
```

```python
# Clean Gender Column
# Standardize similar values (Man -> Male, Woman -> Female, etc.)
df["Gender"] = df["Gender"].replace({
    "Man": "Male",
    "Woman": "Female",
    "Nonbinary": "Non-binary",
    "Non-binary, genderqueer, or gender non-conforming": "Non-binary",
    "Prefer to self-describe": "Other",
    "A different identity": "Other",
    "Unknown": "Prefer not to say"  # Group unknown with prefer not to say
})

# Encode Gender Values
gender_map = {
    "Male": 1,
    "Female": 0,
    "Non-binary": 2,
    "Other": 3,
    "Prefer not to say": -1
}

df["Gender_Encoded"] = df["Gender"].map(gender_map)

# Show a Clean Summary 
print("✅ Gender Column Encoded Successfully!\n")

# Check Result
print("Unique Gender Values After Cleaning & Encoding:")
print(df[["Gender", "Gender_Encoded"]].drop_duplicates())
```

```python
# Export cleaned dataset
df.to_csv("cleaned_kaggle_survey.csv", index=False)
print("✅ Cleaned dataset exported successfully!")
```

# ***Visualisation***

```python
# AGE DISTRIBUTION
plt.figure(figsize=(8,4))
sns.countplot(
    data=df,
    x="Age_Range",
    order=df["Age_Range"].value_counts().index,
    hue="Age_Range",          
    palette="Set2",
    legend=False              
)

plt.title("Distribution of Respondents by Age Range")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Top 10 Programming Languages
plt.figure(figsize=(12,6))
lang_counts = df['Recommended_Language'].value_counts().head(10)
sns.barplot(y=lang_counts.index, x=lang_counts.values, hue=lang_counts.index, palette="viridis", legend=False)
plt.title("Top 10 Programming Languages", fontsize=14)
plt.xlabel("Number of Mentions")
plt.ylabel("Programming Language")
for index, value in enumerate(lang_counts.values):
    plt.text(value + 300, index, str(value), va='center', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Top 10 Jobs Title
plt.figure(figsize=(12,6))
job_counts = df['Job_Title'].value_counts().head(10)
sns.barplot(y=job_counts.index, x=job_counts.values, palette="rocket", hatch="//")
plt.title("Top 10 Job Titles", fontsize=14)
plt.xlabel("Number of Respondents")
plt.ylabel("Job Title")
for index, value in enumerate(job_counts.values):
    plt.text(value + 150, index, str(value), va='center', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
#Top 10 Countries by Respondents
plt.figure(figsize=(12,6))
country_counts = df['Country'].value_counts().head(10)
sns.barplot(y=country_counts.index, x=country_counts.values, hue=country_counts.index, palette="tab10", legend=False)
plt.title("Top 10 Countries by Respondents", fontsize=14)
plt.xlabel("Number of Respondents")
plt.ylabel("Country")
for index, value in enumerate(country_counts.values):
    plt.text(value + 100, index, str(value), va='center', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# EDUCATION DISTRIBUTION
plt.figure(figsize=(10,5))
edu_counts = df["Education"].value_counts()

sns.barplot(
    x=edu_counts.values,
    y=edu_counts.index,
    palette="muted"
)

plt.title("Most Common Education Levels among Respondents")
plt.xlabel("Number of Respondents")
plt.ylabel("Education Level")
plt.tight_layout()
plt.show()
```
